# Q1: What is MongoDB? Explain non-relational databases in short. In which scenarios is it preferred to use MongoDB over SQL databases?

MongoDB is a NoSQL database management system that stores data in flexible, JSON-like documents. It is part of the family of non-relational databases. Non-relational databases, or NoSQL databases, are designed to handle unstructured or semi-structured data and offer flexible schemas. Unlike traditional relational databases (SQL databases), NoSQL databases do not require a fixed schema, and they are often more scalable and suitable for handling large amounts of data with diverse structures.

Scenarios where MongoDB is preferred over SQL databases include:

When dealing with large amounts of data with varying structures.
Projects that require horizontal scalability and distribution of data across multiple servers or clusters.
Agile development environments where schema changes are frequent.
Applications that benefit from the flexibility of a document-based data model (JSON-like documents).
# Q2: State and Explain the features of MongoDB.

Some key features of MongoDB include:

Flexible Schema: MongoDB doesn't require a predefined schema, allowing for dynamic and evolving data models.
Document-Oriented: Data is stored in BSON (Binary JSON)-like documents, which can contain arrays and other documents.
Scalability: MongoDB can scale horizontally by sharding data across multiple servers or nodes.
Indexing: Supports efficient indexing to improve query performance.
Aggregation Framework: Provides powerful aggregation and data processing capabilities.
Replication: Supports replica sets for high availability and data redundancy.
Ad Hoc Queries: Supports rich and expressive queries.
GridFS: A specification for storing large files, such as images or videos.

# Q5: Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

The find() method is used to query data from a MongoDB collection. It returns a cursor, which can be iterated to retrieve documents.

python
Copy code
    # Find documents where age is greater than 30
query = {"age": {"$gt": 30}}
result = mycollection.find(query)

for record in result:
    print(record)

# Q7: Explain why delete_one(), delete_many(), and drop() are used.

delete_one(filter): Deletes a single document that matches the specified filter criteria.

delete_many(filter): Deletes all documents that match the specified filter criteria.

drop(): Drops the entire collection, including all its documents.

These methods are used to remove documents or collections from the database. delete_one() and delete_many() are more targeted, allowing selective removal based on specified criteria. drop() is more drastic, as it removes the entire collection.
